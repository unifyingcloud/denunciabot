 ## TFIDF 

```python
tweet_tokenizer = casual.TweetTokenizer()
spanish_stemmer = SnowballStemmer("spanish")
spanish_stopwords = stopwords.words("spanish")
```

```python
def process_text(txt):
    txt = txt.lower().strip()
    txt = u''.join([c for c in txt if c not in string.punctuation])
    txt = u''.join([c for c in txt if c not in string.digits])
    return [spanish_stemmer.stem(t) for t in tweet_tokenizer.tokenize(txt) if t not in spanish_stopwords and len(t) > 1]
```

```python
df_promotions_es["features_clean"] = df_promotions_es["features"].apply(lambda x: process_text(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn import metrics
from sklearn.cluster import KMeans, MiniBatchKMeans
```

```python
tfidf = TfidfVectorizer(tokenizer=process_text, min_df=2)
tfs = tfidf.fit_transform(df_promotions_es["features"].values)
```

```python
svd = TruncatedSVD(5)
normalizer = Normalizer(copy=False)
lsa = make_pipeline(svd, normalizer)

X = lsa.fit_transform(tfs)
```

```python
km = MiniBatchKMeans(n_clusters=5, init='k-means++', n_init=1,
                     init_size=1000, batch_size=1000, verbose=False)
```

```python
km.fit(X)
```

```python
print("Top terms per cluster:")

original_space_centroids = svd.inverse_transform(km.cluster_centers_)
order_centroids = original_space_centroids.argsort()[:, ::-1]

terms = tfidf.get_feature_names()
for i in range(5):
    print("Cluster %d:" % i, end='')
    for ind in order_centroids[i, :20]:
        print(' %s' % terms[ind], end='')
    print()
```
